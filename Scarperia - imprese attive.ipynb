```python
df_2015 = pd.read_csv(r'/Users/marcodugini/Desktop/open12015.csv', sep=';', encoding='latin1', index_col="COMUNE")
df_2014 = pd.read_csv(r'/Users/marcodugini/Desktop/registrateeattiveopen12014.csv', 
                      sep=';', encoding='latin1', index_col="comune")
```

```python
df_2015 = df_2015.loc["SCARPERIA E SAN PIERO"]
df_2014 = df_2014.loc["SCARPERIA E SAN PIERO"]
```

# Imprese attive nel 2015 nel comune di Scarperia e San piero per natura giuridica e settore Ateco 

```python
df__2015 = df_2015[["anno", "GRUPPO_NATURA_GIURIDICA", "ATECO_DES_SEZIONE", "IMPRESE ATTIVE"]]
df__2015
```

# Totale imprese attive a Scarperia e San Piero per settori Ateco 2015

```python
composizione_ateco_2015 = df__2015.groupby("ATECO_DES_SEZIONE").sum().drop("anno", axis=1)
composizione_ateco_2015
```

# Composizione percentuale settori Ateco 2015

```python
composizione_ateco_2015["TOTALE IMPRESE"] = 975 
(composizione_ateco_2015.divide(composizione_ateco_2015["TOTALE IMPRESE"], axis=0)*100).drop("TOTALE IMPRESE", axis=1)

## I primi tre settori ATECO nel comune di Scarperia e San Piero sono le Costruzioni (23,3%)
# a seguire il Commercio e la riparazione di autoveicoli e motocicli (19,9%)
# e la Manifattura (16,9%)
```

# Variazione percentuale 2015-2014

```python
df__2014 = df_2014[["anno", "natura_giuridica", "settore_attivita", "attive"]]
composizione_ateco_2014 = df__2014.groupby("settore_attivita").sum().drop("anno", axis=1)

raccordo = dict(zip(composizione_ateco_2014.index.unique().tolist(), composizione_ateco_2015.index.unique().tolist()))

def dizionario(x):
  try:
      return raccordo[x]
  except:
      return ""
    
df_2014["settore_attivita"] = df_2014["settore_attivita"].apply(lambda x: dizionario(x))


variazione_percentuale = composizione_ateco_2015["IMPRESE ATTIVE"] / composizione_ateco_2014["attive"] * 100 - 100
variazione_percentuale
```

```python
# Le imprese totali sono aumentate nel 2015 (975) rispetto al 2014 (957)
# In particolare, i settori economici più in crescita sono: le Attività professionali, scientifiche e tecniche (+36,4%), Attività finanziarie e ass. (+9%), Servizi di Informazione e Comunicazione (+8,3%)
```
